##**Response Selection**

```python
def random_selection_percentage(test_data):

    total_samples = len(test_data)
    correct_selections = 0

    for sample in test_data:
        correct_answer = sample[1]
        wrong_answer = sample[2]

        selected_answer = random.choice([correct_answer, wrong_answer])

        if selected_answer == correct_answer:
            correct_selections += 1

    percentage_correct = (correct_selections / total_samples) * 100
    return percentage_correct
```

```python
# This is the result for random baseline withou
test_data = pd.read_pickle('/content/drive/MyDrive/Perspective Paper/Original_Data/final_test.pkl')
percentage_correct = random_selection_percentage(test_data)
print(f"Random Baseline - Percentage Correct: {percentage_correct:.2f}%")
```
